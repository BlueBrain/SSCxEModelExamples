# Validation

This notebook demonstrates the validation step's results and their interpretation.

The validation step is performed on the input morphologies available at `input/morphologies` directory.

The directory contains 1000 morphologies. Running all of them takes hours, therefore the run command is not included in this interactive notebook.

To run them, use the following command.

```Bash
python main.py att_conf.json
```



`att_conf.json` file contains the configuration parameter. For a quicker run, one can reduce the `n_morph` value on the file. 

The output directory is also specified in this file.


````python
from pathlib import Path

import matplotlib.pyplot as plt
%matplotlib inline

from lib import plottools as pt
from plot import plot_results
````

````python
figname = "Attenuation_cADpyr_L5TPC_e3026da"
results_paths = [Path("output") / "results.json"]
````

## Attenuation plots for cADpyr_L5TPC combination

The plots below show how attenuation and amplitude behave when we move farther from the soma.

The plots are enriched with the results of various experimental datasets from the literature for comparison.

````python
fig = pt.make_figure(
    figname=figname, figsize_mm=(200, 160), fontsizes=(12, 14)
)
plot_results(
    fig, figname=figname, results_paths=results_paths, plotwhat=["EPSP", "bAP"], legend_fontsize=8
)

plt.show()
````
